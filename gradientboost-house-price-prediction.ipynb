# Checking Missing Values

```python
df.isnull().sum()
```

```python
df.loc[df['total_bedrooms'].isnull(), 'total_bedrooms']=df['total_bedrooms'].median()
```

The dataset was checked for missing values to ensure data quality. Only the `total_bedrooms` feature contained missing entries, which were handled using median imputation to avoid the influence of outliers.

# Encoding Categorical Variables

```python
df=pd.get_dummies(df, columns=['ocean_proximity'], drop_first=True)
df
```

The categorical feature `ocean_proximity` was converted into numerical form using **one-hot encoding**. This transformation allows the model to process categorical data without introducing any artificial ordering.

# Selecting Target and Feature Variables

```python
x=df.drop(columns=['median_house_value'])
y=df.median_house_value
```

The target variable selected for this regression task is `median_house_value`. All remaining relevant numerical and encoded categorical features were used as input variables for model training.

# Splitting Training and Test Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y,test_size=0.2, random_state=42)
```

# Traing the Model

```python
from sklearn.ensemble import GradientBoostingRegressor
model=GradientBoostingRegressor(n_estimators=300, learning_rate=0.05, max_depth=4)
model.fit(x_train, y_train)
```

A **Gradient Boosting Regressor** was trained on the processed dataset. Hyperparameters such as learning rate, number of estimators, and tree depth were tuned to improve predictive performance while preventing overfitting.

# Model Prediction

```python
y_pred=model.predict(x_test)
```

# Model Evaluation

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
r2=r2_score(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
mae=mean_absolute_error(y_test, y_pred)
rmse=mean_squared_error(y_test, y_pred, squared=False)
print(f'R2 Score={r2}\nMean Sqaured Error= {mse}\nMean Absolute Error= {mae}\nRoot Mean Sqaured Error= {rmse}')
```

Model performance was evaluated using regression metrics including **R² score**, **Mean Absolute Error (MAE)**, and **Root Mean Squared Error (RMSE)**. These metrics provide insights into variance explained and average prediction error.
